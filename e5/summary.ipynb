```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def get_data(file):
    return pd.read_csv(file)

def get_stats(data):
    return pd.Series({
        'x_mean': data.mean()['x'],
        'y_mean': data.mean()['y'],
        'x_std': data.std()['x'],
        'y_std': data.std()['y'],
        'x_range': (data.min()['x'], data.max()['x']),
        'y_range': (data.min()['y'], data.max()['y']),
        'corr': data.corr()['x']['y']
    })
```

```python
data1 = get_data('data-1.csv').dropna(subset=['x', 'y'])
data2 = get_data('data-2.csv').dropna(subset=['x', 'y'])
data3 = get_data('data-3.csv').dropna(subset=['x', 'y'])
data4 = get_data('data-4.csv').dropna(subset=['x', 'y'])
data5 = get_data('data-5.csv').dropna(subset=['x', 'y'])
data6 = get_data('data-6.csv').dropna(subset=['x', 'y'])

data1_stats = get_stats(data1)
data2_stats = get_stats(data2)
data3_stats = get_stats(data3)
data4_stats = get_stats(data4)
data5_stats = get_stats(data5)
data6_stats = get_stats(data6)

all_stats = pd.DataFrame()
all_stats['data_1'] = data1_stats
all_stats['data_2'] = data2_stats
all_stats['data_3'] = data3_stats
all_stats['data_4'] = data4_stats
all_stats['data_5'] = data5_stats
all_stats['data_6'] = data6_stats

all_stats = all_stats.transpose()
```

```python
plt.figure(figsize=(35, 27))

plt.subplot(661)
plt.scatter(data1['x'], data1['y'], c='b')
plt.title('Data Set 1')

plt.subplot(662)
plt.scatter(data2['x'], data2['y'], c='g')
plt.title('Data Set 2')


plt.subplot(663)
plt.scatter(data3['x'], data3['y'], c='y')
plt.title('Data Set 3')

plt.figure(figsize=(35, 27))

plt.subplot(664)
plt.scatter(data4['x'], data4['y'], c='r')
plt.title('Data Set 4')

plt.subplot(665)
plt.scatter(data5['x'], data5['y'], c='m')
plt.title('Data Set 5')

plt.subplot(666)
plt.scatter(data6['x'], data6['y'], c='k')
plt.title('Data Set 6')

plt.show()
```
