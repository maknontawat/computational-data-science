```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_views_data(filename):
    return pd.read_csv(filename, sep=' ', header=None, index_col=1, names=['lang', 'page', 'views', 'bytes'])
```

```python
dataset1 = get_views_data("pagecounts-20190509-120000.txt")
dataset2 = get_views_data("pagecounts-20190509-130000.txt")
```

```python
df1 = pd.DataFrame(dataset1)
df2 = pd.DataFrame(dataset2)
```

```python
def draw_plots(df1, df2):
    plt.figure(figsize=(15, 7))
    
    # sort df1 by views
    sort_by_views = df1.sort_values(by=['views'])
    
    # plots distribution of views for dataset 1
    plt.subplot(1, 2, 1)
    plt.title('Popularity Distribution')
    plt.xlabel('Rank')
    plt.ylabel('Views')
    plt.plot(sort_by_views['views'].values)
    
    # gets the two series into the same DataFrame
    views_df = pd.DataFrame()
    views_df['day1'] = df1['views']
    views_df['day2'] = df2['views']
    
    # plots daily views for dataset 1 and 2
    plt.subplot(1, 2, 2)
    plt.title('Daily Correlation')
    plt.xlabel('Day 1 views')
    plt.ylabel('Day 2 views')
    plt.xscale('log')
    plt.yscale('log')
    plt.scatter(views_df['day1'], views_df['day2'], c='b')
    
    plt.show()

draw_plots(df1, df2)
```

```python
views_df = pd.DataFrame()
views_df['day1'] = df1['views']
views_df['day2'] = df2['views']
plt.xscale('log')
plt.yscale('log')
plt.scatter(views_df['day1'], views_df['day2'], c='b')
```
